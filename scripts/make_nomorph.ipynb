```python
import os
import sys
import glob
from os import listdir
from os.path import isfile, join
import pandas as pd
```

```python
os.getcwd()
```

```python
all_configs = glob.glob("../example/simulated_data-*/*/")
```

```python
config_folder = all_configs[0]
config_folder
```

```python
for config_folder in all_configs:
    onlyfiles = [f for f in listdir(config_folder) if isfile(join(config_folder, f))]
    for file in onlyfiles:
        try:
            os.mkdir(f'{config_folder}/no-morph')
        except:
            pass
        df = pd.read_csv(config_folder + '/'+ file)
        df = df[df['id'].str.contains('l')]
        nomorph_name = f'{config_folder}/no-morph/{file}'
        df.to_csv(nomorph_name, index=False)
        print(nomorph_name)
```
