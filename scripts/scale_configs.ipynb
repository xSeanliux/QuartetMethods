```python
import pandas as pd
import os 
import numpy as np
import glob
```

```python
factors = [0.5, 1.0, 2.0, 4.0, 8.0]
configdir = './QuartetMethods/example/configs'
```

```python
from os import listdir
from os.path import isfile, join
folders = [dirpath for dirpath, _, _ in os.walk(configdir)]
onlyfiles = [f for f in listdir(configdir) if isfile(join(configdir, f))]
onlyfiles
```

```python
def makeConfig(configdir=configdir, factor=1.0):
    # this function scales the evolution rates in the given config files in configdir by factor and saves them elsewhere
    onlyfiles = [f for f in listdir(configdir) if isfile(join(configdir, f))] 
    newdir = f'{configdir}-{factor}'
    try:
        os.mkdir(newdir)
    except:
        pass
    # change evolution factor
    for filename in onlyfiles: 
        with open(configdir+'/'+filename, "r") as f:
            df = pd.read_csv(f)
            df.rename(columns={'Unnamed: 0': ""}, inplace=True)
            arr = np.array(pd.to_numeric(df.iloc[5, 1:])) * factor
            df.iloc[5, 1:] = [str(x) for x in arr]
            new_filename = f'{newdir}/{filename}' 
            df.to_csv(new_filename, index=False)
```

```python
config_folder = all_configs[0]
config_folder
```

```python
for config_folder in all_configs:
    onlyfiles = [f for f in listdir(config_folder) if isfile(join(config_folder, f))]
    for file in onlyfiles:
        try:
            os.mkdir(f'{config_folder}/no-morph')
        except:
            pass
        df = pd.read_csv(config_folder + '/'+ file)
        df = df[df['id'].str.contains('l')]
        nomorph_name = f'{config_folder}/no-morph/{file}'
        df.to_csv(nomorph_name, index=False)
        print(nomorph_name)
```
